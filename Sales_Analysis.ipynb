### Monthly Revenue Over Two-Year Period

```python
data['Month'].head()
```

```python
plt.figure(figsize=(8,6))
plt.plot(data['Month'],data['Revenue'],marker='s',color='red',label='Revenue')
plt.title('Monthly Revenue Over Time',color='blue')
plt.xlabel("Time",color='green')
plt.ylabel("Revenue",color='green')
plt.xticks(rotation=90)
plt.legend()
plt.grid(linestyle='--',alpha=0.7,linewidth=0.5)
plt.show()
```

Key points to note:
1. The revenue shows a consistent upward trend from January 2020 to December 2021.
2. The consistent growth in revenue suggests positive business performance and increasing sales.
3. There is no clear indication of seasonality or cyclic patterns in the revenue, implying steady growth without significant fluctuations.

### Total number of orders per month

```python
plt.figure(figsize=(8,6))
plt.bar(data["Month"],data["Orders"],color='yellow')
plt.title("Total Orders per Month",color='red')
plt.xlabel("Time")
plt.ylabel("Total Orders")
plt.xticks(rotation=90)
plt.grid(linestyle='--',linewidth=0.5)
plt.show()
```

From the above graph, similar to revenue, the number of orders shows a consistent upward trend over the same period.
The increase in orders aligns with the increase in revenue, indicating that the revenue growth is driven by a higher volume of orders.
Like the revenue, there is no strong seasonal pattern, indicating a steady increase in orders.

### Relationship between the number of orders and the average order value.

```python
plt.figure(figsize=(8,6))
plt.scatter(data['Orders'],data["Average_Order_Value"])
plt.title('Relationship Between Number of Orders and Average Order',color='blue')
plt.xlabel("Orders")
plt.ylabel('Average_Order_Value')
plt.grid(linestyle='--')
plt.show()
```

There is a constant Average Order Value. The average order value is constant at $50 across all data points. This means that the growth in revenue is entirely due to the increase in the number of orders. This consistency suggests that the pricing strategy and customer purchasing behavior regarding the value per order have remained unchanged over the analyzed period.

### Conclusion
From the above analysis it is clear that their is a positive growth. Both revenue and the number of orders have shown consistent growth over the analyzed period, indicating a healthy and growing business.
The revenue growth is driven by an increasing number of orders rather than changes in the average order value.
The lack of significant fluctuations or seasonal effects suggests that the business has maintained steady performance without major disruptions or peak seasons.

### Subplot containing all three visualizations created

```python
fig, (ax1, ax2, ax3) = plt.subplots(nrows=3, ncols=1, figsize=(10, 16))

ax1.plot(data['Month'], data['Revenue'],color='red',marker='o',label='Revenue')
ax1.set_title('Monthly Revenue Over Time',fontsize=14)
ax1.set_ylabel('Revenue')
ax1.set_xlabel('Time')
ax1.tick_params(axis='x',rotation=90)
ax1.legend()
ax1.grid(linestyle='--',alpha=0.7,linewidth=0.5)

ax2.bar(data['Month'], data['Orders'],label='Total Orders',color='green')
ax2.set_title('Monthly Orders',fontsize=14)
ax2.set_ylabel('Total Orders')
ax2.set_xlabel('Time')
ax2.tick_params(axis='x',rotation=90)
ax2.legend()
ax2.grid(axis='y',linestyle='--')

ax3.scatter(data['Orders'], data['Average_Order_Value'],label='Order Value',color='orange')
ax3.set_title('Orders vs Average Order Value',fontsize=14)
ax3.set_xlabel('Orders')
ax3.set_ylabel('Average Order Value')
ax3.legend()
ax3.grid(linestyle='--')

plt.tight_layout()
plt.savefig('sales_analysis.png', dpi = 300)
plt.show()
```
